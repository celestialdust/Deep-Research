```python
report = """# Comparative efficacy and safety of zanubrutinib versus pirtobrutinib in relapsed/refractory CLL/SLL (as of November 25, 2025)

## Overview and clinical context
Relapsed/refractory chronic lymphocytic leukemia/small lymphocytic lymphoma (CLL/SLL) has been transformed by Bruton’s tyrosine kinase inhibitors (BTKi). Among adults who have progressed on or after a covalent BTKi—or who are BTKi‑naïve but require a BTKi in the relapsed setting—the key therapeutic choices now include zanubrutinib, a covalent BTKi optimized for on‑target selectivity and exposure, and pirtobrutinib, a highly selective, reversible (non‑covalent) BTKi designed to retain activity against BTK C481 mutations. As the evidence base matured, distinct clinical niches emerged. Zanubrutinib’s phase 3 head‑to‑head data versus ibrutinib in BTKi‑naïve R/R CLL demonstrate superior efficacy and a favorable cardiovascular profile. Pirtobrutinib’s data in covalent BTKi‑pretreated populations—including randomized superiority versus idelalisib+rituximab or bendamustine+rituximab—establish it as the preferred BTK‑targeted therapy after covalent BTKi failure. Across clinically relevant subgroups (prior BTKi/BCL2 exposure, BTK C481 mutation, del[17p]/TP53 abnormality, IGHV status, BTKi intolerance vs resistance), both agents show consistent disease control with differing strengths that inform sequencing.

Direct cross‑trial comparisons are limited by different designs, baseline risks, and follow‑up durations. The synthesis below therefore emphasizes within‑trial comparisons and mechanistic fit, and provides quantitative endpoints where available. Key primary sources include the ALPINE final analysis in Blood for zanubrutinib, pooled cardiovascular safety analyses for zanubrutinib, the BRUIN phase 1/2 NEJM report and JCO phase 3 BRUIN CLL‑321 for pirtobrutinib, and current regulatory labels.

## Mechanisms and dosing
Zanubrutinib is an irreversible covalent BTKi that binds BTK at Cys481, achieving sustained BTK occupancy with fewer off‑target kinase effects than first‑generation ibrutinib. The recommended dose is 160 mg twice daily (or 320 mg once daily) and therapy is continuous until progression or intolerance. Reduced off‑target inhibition (e.g., CSK, TEC, EGFR family) is associated with lower rates of atrial arrhythmias and selected nonhematologic toxicities seen historically with ibrutinib [1,2,4].

Pirtobrutinib is a reversible, non‑covalent BTKi that binds the BTK ATP pocket independent of Cys481, inhibiting both wild‑type and C481‑mutated BTK. The recommended dose is 200 mg once daily; treatment is continuous until progression or intolerance. This design overcomes the dominant covalent BTKi resistance mechanism (BTK C481S/R/Y mutations), while maintaining high BTK selectivity [5,6,7].

Mechanistic implication: In patients with BTK C481‑mutated disease after covalent BTKi exposure, pirtobrutinib retains activity. In BTKi‑naïve R/R disease where a covalent BTKi is indicated, zanubrutinib is favored over ibrutinib for efficacy and cardiovascular safety.

## Trial populations and design considerations
The phase 3 ALPINE trial (NCT03734016) randomized 652 BTKi‑naïve adults with R/R CLL/SLL to zanubrutinib versus ibrutinib, with a median follow‑up of 42.5 months at the final analysis. No prior BTKi exposure was permitted; there was no crossover. Endpoints included ORR, PFS, OS, and prespecified safety analyses (notably atrial fibrillation/flutter), with subgroup analysis for del[17p]/TP53 [1].

Pirtobrutinib’s phase 1/2 BRUIN trial (NCT03740529) established monotherapy activity in advanced B‑cell malignancies; the key CLL/SLL cohort comprised covalent BTKi‑pretreated patients with or without prior venetoclax. Efficacy was assessed per 2018 iwCLL (including PR with lymphocytosis). Additional analyses reported outcomes by BTK C481 mutation, del[17p]/TP53, IGHV status, PLCG2 mutations, and an intolerance cohort [5].

The phase 3 BRUIN CLL‑321 (NCT04666038) randomized 238 adults previously treated with a covalent BTKi to pirtobrutinib versus investigator’s choice of idelalisib+rituximab or bendamustine+rituximab (IdelaR/BR). Crossover to pirtobrutinib was allowed upon progression. The primary endpoint was PFS per 2018 iwCLL by independent review; prespecified subgroup analyses included prior venetoclax, del[17p]/TP53, IGHV status, and complex karyotype [6,10].

Design differences that constrain direct comparison include BTKi‑naïve versus BTKi‑pretreated populations, distinct control arms (ibrutinib versus IdelaR/BR), allowance for crossover (CLL‑321), and COVID‑19 era effects on infection profiles. Follow‑up durations and assessment schedules also differ.

## Efficacy endpoints: quantitative synthesis
Across endpoints, pirtobrutinib’s strength is in covalent BTKi‑pretreated disease; zanubrutinib’s strength is in BTKi‑naïve R/R disease versus ibrutinib.

### Overall response rate (ORR, including PR‑L per 2018 iwCLL) and CR/CRi
In ALPINE, zanubrutinib achieved higher ORR than ibrutinib in BTKi‑naïve R/R CLL/SLL: ORR 85.6% versus 75.4%, with CR/CRi 11.6% versus 7.7%. Including PR‑L or better, rates were 90.2% versus 82.8%. Responses deepened over time [1].

In BRUIN phase 1/2 (covalent BTKi‑pretreated CLL/SLL), pirtobrutinib’s ORR was 73.3% (CR 1.6%), and when PR‑L was included, 82.2%. In an expanded dataset, ORR including PR‑L was 81.6% with CR ~1.8% [5]. In the randomized BRUIN CLL‑321, pirtobrutinib’s ORR including PR‑L was 69% versus 50% with IdelaR/BR [6].

Interpretation: Zanubrutinib shows high ORR and greater CR/CRi in BTKi‑naïve R/R disease versus ibrutinib. In the covalent BTKi‑pretreated setting, pirtobrutinib maintains high ORR—including PR‑L—and improves ORR versus IdelaR/BR.

### Progression‑free survival (PFS)
In ALPINE, zanubrutinib significantly improved PFS versus ibrutinib: HR 0.68 (95% CI, 0.54–0.84). The 36‑month PFS rates were 65.4% versus 54.4%. In del[17p]/TP53‑abnormal disease, the PFS HR was 0.51 (95% CI, 0.33–0.78). Median PFS was not reported; OS remained not reached in both arms [1].

In BRUIN phase 1/2 (post‑covalent BTKi), pirtobrutinib’s median PFS was 19.6 months overall; 22.1 months in patients without prior venetoclax; 16.8 months after prior BTKi plus BCL2 inhibitor; and approximately 17.5–20.0 months regardless of BTK C481 mutation status [5]. Median PFS in del[17p]/TP53‑abnormal and unmutated IGHV subgroups was 16.9 and 18.7 months, respectively [5].

In BRUIN CLL‑321, pirtobrutinib significantly improved PFS versus IdelaR/BR: independent review HR 0.54 (95% CI, 0.39–0.75), with medians 14.0 versus 8.7 months; investigator assessment HR 0.48 (95% CI, 0.34–0.67), 15.3 versus 9.2 months. PFS benefit was consistent in key subgroups: del[17p]/TP53 (HR 0.59), unmutated IGHV (HR 0.61), complex karyotype (HR 0.37), prior venetoclax (HR 0.54) [6].

Interpretation: In BTKi‑naïve R/R disease, zanubrutinib delivers superior PFS versus ibrutinib, with enhanced benefit in del[17p]/TP53. In BTKi‑pretreated disease, pirtobrutinib produces clinically meaningful PFS with randomized superiority versus IdelaR/BR and robust performance across high‑risk subgroups.

### Overall survival (OS)
ALPINE showed a favorable OS trend for zanubrutinib versus ibrutinib (HR 0.77; 95% CI, 0.55–1.06) with median OS not reached [1]. In BRUIN CLL‑321, OS was confounded by crossover (76% of eligible IC patients crossed to pirtobrutinib): unadjusted HR 1.09 (95% CI, 0.68–1.75), with adjusted sensitivity analyses not statistically significant [6].

Interpretation: OS differences are immature and/or confounded; available evidence shows no OS detriment with either agent in their respective settings.

### Duration of response (DOR)
ALPINE reports response deepening but not a specific median DOR for zanubrutinib in the final paper [1]. In BRUIN monotherapy, DOR paralleled PFS durability, with sustained responses beyond 12–18 months in many subgroups [5]. Direct quantitative DOR comparisons across trials are limited by design differences.

### Time‑to‑next‑treatment (TTNT)
ALPINE did not report TTNT. BRUIN CLL‑321 showed a marked TTNT benefit with pirtobrutinib: HR 0.37 (95% CI, 0.25–0.52), median 24.0 months versus 10.9 months. TTNT was longer in venetoclax‑naïve patients (29.5 vs 12.5 months; HR 0.36) and remained favorable after prior venetoclax (20.0 vs 8.7 months; HR 0.37) [6].

Interpretation: In the post‑covalent BTKi setting, pirtobrutinib meaningfully delays subsequent therapy, which is clinically relevant for quality of life and resource planning.

### Measurable residual disease (MRD)
ALPINE did not report MRD. Pirtobrutinib monotherapy MRD data are limited; however, early combination cohorts (pirtobrutinib + venetoclax ± anti‑CD20) have reported high rates of undetectable MRD during fixed‑duration therapy, suggesting potential for time‑limited deep remissions; these data are preliminary and outside the core monotherapy comparisons [14].

## Safety endpoints: quantitative synthesis
Safety reflects molecular selectivity, patient baseline risks, trial era, and exposure duration. Both agents require vigilance for infections, bleeding, and cytopenias; cardiovascular safety is a differentiator.

### Atrial fibrillation/flutter, other cardiac events, and cardiac deaths
In ALPINE, zanubrutinib significantly reduced atrial fibrillation/flutter versus ibrutinib (7.1% vs 17.0%), with fewer overall cardiac AEs (25.9% vs 35.5%) and zero cardiac‑related deaths versus six on ibrutinib [1]. Head‑to‑head pooled analyses (ASPEN + ALPINE) corroborated reduced AF/flutter with zanubrutinib versus ibrutinib (6.1% vs 15.6%; exposure‑adjusted incidence rate [EAIR] ratio ~0.31) and lower cardiac deaths (0.2% vs 1.7%) [2]. A larger pooled monotherapy analysis also showed low AF/flutter incidence (4.8%; grade ≥3, 2.0%) and rare ventricular arrhythmias (0.7% symptomatic; grade ≥3, 0.4%) [2,3].

In BRUIN CLL‑321, pirtobrutinib had low AF/flutter incidence (2.6%; two cases had prior AF on covalent BTKi) and low hypertension rates; grade ≥3 cardiac events were uncommon, and exposure‑adjusted rates favored pirtobrutinib versus IdelaR/BR [6]. The US label reports AF/flutter ~3.2% overall (grade ≥3, 1.5%) across pooled datasets, with very low rates of serious ventricular arrhythmias [7].

Interpretation: Both agents have low atrial arrhythmia rates relative to historical ibrutinib experience. Zanubrutinib’s head‑to‑head data substantiate a substantial reduction in AF/flutter and cardiac deaths versus ibrutinib. Pirtobrutinib shows similarly low AF/flutter in BTKi‑pretreated populations, including patients with prior AF.

### Hypertension
In ALPINE, hypertension was similar between zanubrutinib and ibrutinib (any grade 27.2% vs 25.3%; grade ≥3 17.0% vs 16.0%), yet cardiac discontinuations and deaths were markedly lower with zanubrutinib [1]. Pooled analyses showed lower or comparable exposure‑adjusted hypertension rates with zanubrutinib and minimal discontinuations due to hypertension [2,3]. In BRUIN CLL‑321, pirtobrutinib had lower any‑grade hypertension (6.9%; grade ≥3, 2.6%) than chemotherapy/PI3Ki comparator regimens [6].

### Infections (grade ≥3 and fatal)
Infections were frequent in ALPINE, in part due to the COVID‑19 era; COVID‑19 infection occurred in 46.0% on zanubrutinib versus 33.3% on ibrutinib, with similar COVID‑19‑related death rates across arms. Grade ≥3 pneumonia was numerically lower with zanubrutinib (7.7% vs 10.5%), and concurrent grade ≥3 neutropenia plus grade ≥3 infection was less frequent (8.3% vs 12.6%) [1]. The US label pooled safety for zanubrutinib reports grade ≥3 infections ~26% and fatal infections ~3% across indications; clinicians should consider prophylaxis in high‑risk patients [4].

For pirtobrutinib, pooled safety shows grade ≥3 infections in ~24% overall, with higher rates in CLL/SLL subpopulations (grade ≥3 ~32%; fatal ~8%) reflective of heavy pretreatment and comorbidity. In BRUIN CLL‑321, exposure‑adjusted TEAE rates favored pirtobrutinib versus IdelaR/BR; pneumonia was common but manageable, and discontinuations due to infections were less frequent with pirtobrutinib [6,7].

Interpretation: Infection risk warrants vigilant monitoring and vaccination strategies for both drugs, especially in older or heavily pretreated patients and during respiratory virus seasons.

### Bleeding/hemorrhage
Class‑typical low‑grade bleeding is common with BTKi. In ALPINE, any‑grade bleeding was similar for zanubrutinib and ibrutinib (43.8% vs 44.4%), with major hemorrhage 4.0% vs 4.9% [1]. Labels advise perioperative holds and caution with antiplatelet/anticoagulants [4]. In BRUIN CLL‑321, pirtobrutinib had mostly low‑grade bleeding with three grade 3 hemorrhages; the control arm had one fatal hemorrhage [6]. The pirtobrutinib label reports major hemorrhage ~3% and fatal bleeding ~0.3% across pooled datasets [7].

### Cytopenias
In ALPINE, neutropenia rates were similar between zanubrutinib and ibrutinib (any grade ~31.5% vs 29.6%; grade ≥3 ~22.8% each), with slightly lower rates of concurrent severe infection plus neutropenia on zanubrutinib [1]. Across pooled datasets, zanubrutinib shows grade 3–4 neutropenia ~21% and thrombocytopenia ~8% [4]. In BRUIN CLL‑321, pirtobrutinib had lower grade ≥3 neutropenia versus IdelaR/BR (20.7% vs 27.5%), and exposure‑adjusted incidence ratios favored pirtobrutinib for anemia and neutropenia [6].

### Diarrhea
Diarrhea was less frequent with zanubrutinib than ibrutinib in ALPINE (18.8% vs 25.6%) [1]. In BRUIN CLL‑321, pirtobrutinib had lower diarrhea rates than IdelaR/BR (15.5% vs 31.2%) [6].

### Dose modifications, discontinuations, and exposure
In ALPINE, median exposure was longer with zanubrutinib (41.2 months vs 37.8 months). Dose interruptions were similar (59.3% vs 62.0%), but dose reductions (14.8% vs 18.2%) and discontinuations due to adverse events (21.4% vs 28.3%) were lower for zanubrutinib; discontinuations due to cardiac AEs were rare (0.9% vs 4.9%) [1]. Pooled analyses across trials corroborate fewer discontinuations and deaths and more patients remaining on zanubrutinib compared with ibrutinib [2,3].

In BRUIN CLL‑321, pirtobrutinib had fewer discontinuations due to AEs than IdelaR/BR (17.2% vs 34.9%) and fewer dose reductions (11.2% vs 36.7%). Median treatment duration on pirtobrutinib (approximately 15 months) exceeded that on IdelaR/BR, consistent with PFS and TTNT benefits [6].

## Subgroup analyses
### Prior covalent BTKi and/or BCL2 inhibitor exposure
ALPINE excluded prior BTKi exposure and therefore informs BTKi‑naïve R/R practice [1]. BRUIN phase 1/2 showed pirtobrutinib retains high activity after covalent BTKi, with median PFS 22.1 months in BTKi‑pretreated, venetoclax‑naïve patients, and 16.8 months after prior venetoclax; heavily pretreated patients (BTKi + BCL2i + PI3Ki + chemoimmunotherapy) had shorter PFS (~13.8 months) [5]. BRUIN CLL‑321 confirmed superiority versus IdelaR/BR across prior venetoclax strata [6].

### BTK C481 mutation
Pirtobrutinib retained activity regardless of BTK C481 mutation status in BRUIN, with similar PFS (~17.5–20 months) and robust ORR including PR‑L. PLCG2 mutations were associated with lower ORR (~56% including PR‑L) [5]. As a covalent BTKi, zanubrutinib’s activity is compromised by BTK C481 mutations; ALPINE did not enroll BTKi‑pretreated patients, and clinical use post‑C481 mutation is mechanistically unfavorable [1].

### del(17p)/TP53 abnormality
In ALPINE, zanubrutinib maintained PFS benefit versus ibrutinib in del[17p]/TP53‑abnormal disease (HR 0.51) [1]. BRUIN CLL‑321 showed pirtobrutinib PFS benefits in del[17p]/TP53‑abnormal subgroups (HR 0.59), indicating consistent performance in genomically high‑risk disease [6].

### IGHV mutation status and complex karyotype
ALPINE did not report IGHV stratification in the final analysis [1]. BRUIN CLL‑321 demonstrated pirtobrutinib PFS benefit in unmutated IGHV (HR 0.61) and complex karyotype (HR 0.37), supporting efficacy across adverse biology [6].

### BTKi intolerance versus resistance
A dedicated BRUIN intolerance cohort showed that pirtobrutinib is well tolerated in BTKi‑intolerant patients, including those who discontinued covalent BTKi for cardiac AEs. The ORR was ~77% and median PFS ~28.4 months; notably, no patient discontinued pirtobrutinib for the same AE that led to prior BTKi discontinuation, and 75% had no recurrence of the same intolerance [13]. ALPINE did not include BTKi‑intolerant rescue cohorts; however, zanubrutinib reduced AF/flutter versus ibrutinib among BTKi‑naïve R/R patients [1].

## Sequencing after covalent BTKi and venetoclax
After covalent BTKi failure (including BTK C481 mutations), pirtobrutinib is the preferred BTK‑targeted therapy based on mechanism and phase 3 superiority versus IdelaR/BR, with consistent benefits across del[17p]/TP53, unmutated IGHV, complex karyotype, and prior venetoclax subgroups. The TTNT advantage indicates meaningful delay of subsequent therapy in both venetoclax‑naïve and venetoclax‑exposed patients, with a longer delay in venetoclax‑naïve disease [6]. Pirtobrutinib may serve as a bridge to time‑limited venetoclax‑based combinations to achieve deep remissions and potentially undetectable MRD, or to cellular therapies in suitable candidates; emerging phase 3 data in treatment‑naïve disease (CLL‑313/314) support broader use but await peer‑reviewed publication for detailed interpretation [11,14].

In BTKi‑naïve R/R patients where a covalent BTKi is indicated, zanubrutinib is favored over ibrutinib due to superior ORR, PFS, and markedly lower atrial fibrillation/flutter and cardiac deaths; this advantage holds in del[17p]/TP53‑abnormal disease [1]. For patients who are BTKi‑intolerant to a covalent agent, pirtobrutinib is an effective and often better tolerated alternative with low recurrence of the same intolerance [13].

## Practical safety management and label guidance
For zanubrutinib, monitor for hemorrhage, infections (including opportunistic), cytopenias, second primary malignancies, and atrial/ventricular arrhythmias; adjust for drug interactions (CYP3A), recommend perioperative holds (3–7 days), and manage hypertension and dermatologic events proactively [4]. For pirtobrutinib, monitor for infections (including opportunistic and severe pneumonia), hemorrhage, cytopenias, atrial fibrillation/flutter, second primary malignancies, and potential hepatotoxicity; avoid warfarin, manage strong CYP3A interactions, and consider dose modifications per label in the event of grade ≥3 toxicities [7]. Vaccination and antimicrobial prophylaxis should be individualized, particularly in older or heavily pretreated patients.

## Limitations of cross‑trial comparisons
Populations differ substantially: ALPINE enrolled BTKi‑naïve R/R patients, whereas BRUIN CLL‑321 enrolled exclusively BTKi‑pretreated patients with high‑risk genomic features. Comparators vary (ibrutinib vs IdelaR/BR), crossover in CLL‑321 confounds OS, and COVID‑19 era effects influence infection patterns. Imaging schedules, response assessment timing, and censoring rules differ, limiting direct comparisons of ORR, DOR, PFS medians, and safety rates. Conclusions therefore emphasize within‑trial effects and mechanistic suitability rather than asserting cross‑agent superiority.

## Summary tables

### Table 1. Comparative efficacy (within‑trial) and key safety signals

| Domain | Zanubrutinib (ALPINE; BTKi‑naïve R/R) | Pirtobrutinib (BRUIN CLL‑321; post‑covalent BTKi) |
| --- | --- | --- |
| ORR (incl. PR‑L) | 90.2% vs 82.8% (vs ibrutinib) [1] | 69% vs 50% (vs IdelaR/BR) [6] |
| CR/CRi | 11.6% vs 7.7% [1] | Low single‑digit CR; responses predominantly PR/PR‑L [6] |
| PFS (HR; median) | HR 0.68 (0.54–0.84); 36‑mo PFS 65.4% vs 54.4% [1] | HR 0.54 (0.39–0.75); 14.0 vs 8.7 mo (IRC) [6] |
| PFS—del[17p]/TP53 | HR 0.51 (0.33–0.78) [1] | HR 0.59 (0.38–0.92) [6] |
| PFS—IGHV unmutated | Not reported | HR 0.61 (0.42–0.88) [6] |
| PFS—complex karyotype | Not reported | HR 0.37 (0.23–0.58) [6] |
| OS | HR 0.77 (0.55–1.06); NR [1] | HR 1.09 (0.68–1.75); crossover allowed [6] |
| TTNT | Not reported | HR 0.37 (0.25–0.52); 24.0 vs 10.9 mo [6] |
| BTK C481 mutation | Not applicable (BTKi‑naïve) | Activity retained; PFS ~17.5–20 mo in phase 1/2 [5] |
| AF/flutter | 7.1% vs 17.0%; cardiac deaths 0 vs 6 [1] | 2.6%; low grade; two with prior AF [6] |
| Hypertension (any; ≥G3) | 27.2%; 17.0% [1] | 6.9%; 2.6% [6] |
| ≥G3 infections | ~26% pooled; pneumonia ≥G3 7.7% in ALPINE [1,4] | ~32% in CLL/SLL pooled; exposure‑adjusted lower vs IdelaR/BR [6,7] |
| Bleeding (major) | 4.0% vs 4.9% [1] | Major ~3% pooled; three ≥G3 events in CLL‑321 [6,7] |
| Neutropenia (≥G3) | 22.8% [1] | 20.7% vs 27.5% (control) [6] |
| Diarrhea (any grade) | 18.8% vs 25.6% [1] | 15.5% vs 31.2% [6] |
| Discontinuations due to AEs | 21.4% vs 28.3% [1] | 17.2% vs 34.9% [6] |
| Dose reductions | 14.8% vs 18.2% [1] | 11.2% vs 36.7% [6] |
| Median exposure | 41.2 vs 37.8 months [1] | ~15 months vs ~5–7 months [6] |

### Table 2. Subgroup mapping: prior therapies/genomics and outcomes

| Subgroup | Zanubrutinib (ALPINE) | Pirtobrutinib (BRUIN phase 1/2 and CLL‑321) |
| --- | --- | --- |
| Prior BTKi exposure | Excluded [1] | Active post‑BTKi: ORR incl. PR‑L ~82%; median PFS 19.6 mo overall [5] |
| Prior BCL2 inhibitor | Not specified | Median PFS 16.8 mo (prior BTKi + BCL2i); TTNT benefit maintained [5,6] |
| BTK C481 mutation | Not reported | Activity retained; PFS ~17.5–20.0 mo; ORR incl. PR‑L high; PLCG2 mutations lower ORR (~56%) [5] |
| del[17p]/TP53 | PFS HR 0.51 vs ibrutinib [1] | PFS HR 0.59 vs IdelaR/BR [6] |
| IGHV unmutated | Not reported | PFS HR 0.61 vs IdelaR/BR [6] |
| Complex karyotype | Not reported | PFS HR 0.37 vs IdelaR/BR [6] |
| BTKi intolerance | Not a rescue cohort | ORR 76.9%; median PFS 28.4 mo; low recurrence of same intolerance [13] |

## Conclusion
For BTK‑pretreated and double‑refractory CLL/SLL, pirtobrutinib offers superior disease control versus idelalisib+rituximab or bendamustine+rituximab and retains activity in BTK C481‑mutated disease, with consistent benefits across del[17p]/TP53, unmutated IGHV, and complex karyotype. It meaningfully prolongs time to next treatment and has low rates of atrial arrhythmias and manageable cytopenias and infections, making it the preferred BTK‑targeted therapy after covalent BTKi failure [5,6]. In BTKi‑intolerant patients, pirtobrutinib provides effective disease control with low recurrence of the inciting intolerance [13].

For BTKi‑naïve R/R CLL/SLL, zanubrutinib demonstrates higher response rates, superior PFS (including in del[17p]/TP53), and markedly fewer atrial fibrillation/flutter events and cardiac deaths than ibrutinib, with longer treatment persistence and fewer discontinuations. These data support zanubrutinib as a preferred covalent BTKi over ibrutinib in this setting [1–3].

Sequencing implications are mechanistically straightforward: after covalent BTKi progression—especially with BTK C481 mutations—use pirtobrutinib; consider transitioning to time‑limited venetoclax‑based combinations or cellular therapies as appropriate. In BTKi‑naïve relapsed patients where a covalent BTKi is indicated, use zanubrutinib rather than ibrutinib for efficacy and cardiovascular safety. Ongoing phase 3 trials (CLL‑313/314) will clarify pirtobrutinib’s role in earlier lines, but current peer‑reviewed evidence already delineates complementary roles for zanubrutinib and pirtobrutinib across the relapse landscape.

### Sources
[1] ALPINE final analysis (Blood, 2024): https://pmc.ncbi.nlm.nih.gov/articles/PMC11738031/  
[2] Cardiovascular events with zanubrutinib vs ibrutinib (Blood Advances, 2024): https://www.sciencedirect.com/science/article/pii/S2473952924001526  
[3] Pooled zanubrutinib monotherapy safety (Blood Advances, 2022): https://www.sciencedirect.com/science/article/pii/S2473952921007746  
[4] BRUKINSA (zanubrutinib) US Prescribing Information (Jan 2025): https://www.accessdata.fda.gov/drugsatfda_docs/label/2025/213217s015lbl.pdf  
[5] BRUIN phase 1/2 pirtobrutinib in CLL/SLL after covalent BTKi (NEJM, 2023): https://www.nejm.org/doi/full/10.1056/NEJMoa2300696  
[6] BRUIN CLL‑321 randomized phase 3 (JCO, 2025): https://ascopubs.org/doi/10.1200/JCO-25-00166  
[7] JAYPIRCA (pirtobrutinib) US Prescribing Information (Dec 2023): https://www.accessdata.fda.gov/drugsatfda_docs/label/2023/216059s001lbl.pdf  
[8] Brukinsa EMA EPAR: https://www.ema.europa.eu/en/medicines/human/EPAR/brukinsa  
[9] Jaypirca EMA EPAR: https://www.ema.europa.eu/en/medicines/human/EPAR/jaypirca  
[10] BRUIN CLL‑321 ASH 2024 abstract: https://ash.confex.com/ash/2024/webprogram/Paper198147.html  
[11] Lilly ASH 2025 press release (BRUIN CLL‑313/314): https://investor.lilly.com/news-releases/news-release-details/lilly-present-data-two-positive-phase-3-studies-jaypirca  
[12] MHRA conditional approval for pirtobrutinib (PharmaTimes, Sept 2025): https://pharmatimes.com/news/lilly-receives-uk-approval-for-pirtobrutinib-in-relapsed-blood-cancers/  
[13] BRUIN intolerance cohort (Haematologica, 2025): https://pubmed.ncbi.nlm.nih.gov/39363864/  
[14] Review of pirtobrutinib combinations and intolerance (Future Oncology, 2025): https://www.tandfonline.com/doi/full/10.1080/14796694.2025.2567834"""
```
